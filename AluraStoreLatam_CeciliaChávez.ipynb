# 2. Ventas por categoría

```python
tienda_categoria = convertir_a_lista(tienda, 'Categoría del Producto')
tienda2_categoria = convertir_a_lista(tienda2, 'Categoría del Producto')
tienda3_categoria = convertir_a_lista(tienda3, 'Categoría del Producto')
tienda4_categoria = convertir_a_lista(tienda4, 'Categoría del Producto')

print(tienda_categoria)
print(tienda2_categoria)
print(tienda3_categoria)
print(tienda4_categoria)
```

```python
def por_categoria(lista_categoria):
  cantidad_categoria = {}

  for categoria in lista_categoria:
    if categoria in cantidad_categoria:
      cantidad_categoria[categoria] += 1
    else:
      cantidad_categoria[categoria] = 1

  return cantidad_categoria
```

```python
cantidad_categoria = por_categoria(tienda_categoria)
cantidad_categoria2 = por_categoria(tienda2_categoria)
cantidad_categoria3 = por_categoria(tienda3_categoria)
cantidad_categoria4 = por_categoria(tienda4_categoria)

print(cantidad_categoria)
print(cantidad_categoria2)
print(cantidad_categoria3)
print(cantidad_categoria4)

llave_maximo = max(cantidad_categoria, key=lambda x: cantidad_categoria[x])
llave_maximo2 = max(cantidad_categoria2, key=lambda x: cantidad_categoria2[x])
llave_maximo3 = max(cantidad_categoria3, key=lambda x: cantidad_categoria3[x])
llave_maximo4 = max(cantidad_categoria4, key=lambda x: cantidad_categoria4[x])

valor_maximo = max(cantidad_categoria.values())
valor_maximo2 = max(cantidad_categoria2.values())
valor_maximo3 = max(cantidad_categoria3.values())
valor_maximo4 = max(cantidad_categoria4.values())
```

```python
print("La categoría donde se encuentran los productos más vendidos de la tienda 1 es:",llave_maximo, "con", valor_maximo, "productos vendidos")
print("La categoría donde se encuentran los productos más vendidos de la tienda 2 es:",llave_maximo2, "con", valor_maximo2, "productos vendidos")
print("La categoría donde se encuentran los productos más vendidos de la tienda 3 es:",llave_maximo3, "con", valor_maximo3, "productos vendidos")
print("La categoría donde se encuentran los productos más vendidos de la tienda 4 es:",llave_maximo4, "con", valor_maximo4, "productos vendidos")
```

Nota #3:
  La función por_categoria recorre todos los elementos de la lista, si un elemento se encuentra por primera vez en la lista, lo añade como llave en un diccionario, ya si se encuentra más de una vez, incrementa el valor de dicha llave. Generando así un diccionario con las categorías como llaves y con la cantidad de productos vendidos por dicha categoría como sus valores.

Nota #4:
  Para poder imprimir la categoría donde se encuentran los productos más vendidos se requirió una función lambda que utiliza el método max() para ubicar el valor más alto del diccionario y acceder mediante su índice. Dicha función se almacena en una variable.

# 3. Calificación promedio de la tienda


```python
def obtener_promedio(lista):
  suma = sum(lista)
  promedio = suma / len(lista)
  return promedio
```

```python
lista_satisfaccion = convertir_a_lista(tienda, 'Calificación')
lista2_satisfaccion = convertir_a_lista(tienda2, 'Calificación')
lista3_satisfaccion = convertir_a_lista(tienda3, 'Calificación')
lista4_satisfaccion = convertir_a_lista(tienda4, 'Calificación')

print(lista_satisfaccion)
print(lista2_satisfaccion)
print(lista3_satisfaccion)
print(lista4_satisfaccion)
```

```python
tienda_satisfaccion = round(obtener_promedio(lista_satisfaccion), 2)
tienda2_satisfaccion = round(obtener_promedio(lista2_satisfaccion), 2)
tienda3_satisfaccion = round(obtener_promedio(lista3_satisfaccion), 2)
tienda4_satisfaccion = round(obtener_promedio(lista4_satisfaccion), 2)

print("La calificación promedio de la tienda 1 es:",tienda_satisfaccion)
print("La calificación promedio de la tienda 2 es:",tienda2_satisfaccion)
print("La calificación promedio de la tienda 3 es:",tienda3_satisfaccion)
print("La calificación promedio de la tienda 4 es:",tienda4_satisfaccion)
```

Nota #5:
  Para calcular el promedio de satisfacción del cliente, se utilizó la función obtener_promedio.
  Suma todos los elementos de la lista y almacena en una variable el cociente de dicha suma entre la longitud de la lista procesada.

# 4. Productos más y menos vendidos

```python
def obtener_productos_por_cantidad(lista):
  productos_por_cantidad = {}

  for producto in lista:
    if producto in productos_por_cantidad:
      productos_por_cantidad[producto] += 1
    else:
      productos_por_cantidad[producto] = 1

  return productos_por_cantidad
```

```python
lista_de_productos = convertir_a_lista(tienda, 'Producto')
lista2_de_productos = convertir_a_lista(tienda2, 'Producto')
lista3_de_productos = convertir_a_lista(tienda3, 'Producto')
lista4_de_productos = convertir_a_lista(tienda4, 'Producto')

print(lista_de_productos)
print(lista2_de_productos)
print(lista3_de_productos)
print(lista4_de_productos)
```

```python
tienda_productos = obtener_productos_por_cantidad(lista_de_productos)
tienda2_productos = obtener_productos_por_cantidad(lista2_de_productos)
tienda3_productos = obtener_productos_por_cantidad(lista3_de_productos)
tienda4_productos = obtener_productos_por_cantidad(lista4_de_productos)

print(tienda_productos)
print(tienda2_productos)
print(tienda3_productos)
print(tienda4_productos)
```

```python
producto_máximo = max(tienda_productos, key=lambda x: tienda_productos[x])
producto_máximo1 = max(tienda2_productos, key=lambda x: tienda2_productos[x])
producto_máximo2 = max(tienda3_productos, key=lambda x: tienda3_productos[x])
producto_máximo3 = max(tienda4_productos, key=lambda x: tienda4_productos[x])

producto_mínimo = min(tienda_productos, key=lambda x: tienda_productos[x])
producto_mínimo2 = min(tienda2_productos, key=lambda x: tienda2_productos[x])
producto_mínimo3 = min(tienda3_productos, key=lambda x: tienda3_productos[x])
producto_mínimo4 = min(tienda4_productos, key=lambda x: tienda4_productos[x])

cantidad_del_producto_más_vendido = max(tienda_productos.values())
cantidad_del_producto_más_vendido2 = max(tienda2_productos.values())
cantidad_del_producto_más_vendido3 = max(tienda3_productos.values())
cantidad_del_producto_más_vendido4 = max(tienda4_productos.values())

cantidad_del_producto_menos_vendido = min(tienda_productos.values())
cantidad_del_producto_menos_vendido2 = min(tienda2_productos.values())
cantidad_del_producto_menos_vendido3 = min(tienda3_productos.values())
cantidad_del_producto_menos_vendido4 = min(tienda4_productos.values())

print("El producto más vendido es",producto_máximo,"con",cantidad_del_producto_más_vendido,"unidades y el producto menos vendido es",producto_mínimo,"con",cantidad_del_producto_menos_vendido,"unidades de la tienda 1.")
print("El producto más vendido es",producto_máximo1,"con",cantidad_del_producto_más_vendido2,"unidades y el producto menos vendido es",producto_mínimo2,"con",cantidad_del_producto_menos_vendido2,"unidades de la tienda 2.")
print("El producto más vendido es",producto_máximo2,"con",cantidad_del_producto_más_vendido3,"unidades y el producto menos vendido es",producto_mínimo3,"con",cantidad_del_producto_menos_vendido3,"unidades de la tienda 3.")
print("El producto más vendido es",producto_máximo3,"con",cantidad_del_producto_más_vendido4,"unidades y el producto menos vendido es",producto_mínimo4,"con",cantidad_del_producto_menos_vendido4,"unidades de la tienda 4.")
```
